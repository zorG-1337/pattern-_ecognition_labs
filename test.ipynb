```python
import matplotlib.pyplot as plt
import cv2

def show_class_diagram(image_path):
    """
    Функция для отображения диаграммы классов из файла изображения.
    """
    # Чтение изображения
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(f"Файл {image_path} не найден.")
    
    # Конвертация изображения из BGR в RGB (OpenCV читает изображения в формате BGR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Отображение изображения с помощью matplotlib
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis("off")  # Отключаем оси для визуализации
    plt.title("Диаграмма классов")
    plt.show()

# Укажите путь к изображению confusion_matrix_normalized.png
image_path = "runs/detect/train8/confusion_matrix_normalized.png"

# Отобразим диаграмму
show_class_diagram(image_path)
```
